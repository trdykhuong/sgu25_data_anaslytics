## 9. Data Cleaning and Preparation

```python
# Handling missing data
float_data = pd.Series([1.2, -3.5, np.nan, 0])
print(float_data)
print(float_data.isna())

string_data = pd.Series(["aardvark", np.nan, None, "avocado"])
print(string_data)
print(string_data.isna())

# Filtering missing data
data = pd.Series([1, np.nan, 3.5, np.nan, 7])
print(data.dropna())
print(data[data.notna()])

# DataFrame with missing data
data = pd.DataFrame([[1.0, np.nan, 2], [np.nan, 3, 4], [np.nan, np.nan, np.nan], [np.nan, 6, 7], [np.nan, 9, np.nan]])
print(data)
print(data.dropna())
print(data.dropna(how="all"))

# Drop columns with all NaN
data[3] = np.nan
print(data.dropna(axis="columns", how="all"))

# Threshold for dropping
df = pd.DataFrame(np.random.standard_normal((7, 3)))
df.iloc[:4, 1] = np.nan
df.iloc[:2, 2] = np.nan
print(df)
print(df.dropna(thresh=2))
```

```python
# Filling missing data
print(df.fillna(0))
print(df.fillna({1: 0.5, 2: 0}))

df = pd.DataFrame(np.random.standard_normal((6, 3)))
df.iloc[2:, 1] = np.nan
df.iloc[4:, 2] = np.nan
print(df)
print(df.ffill())
print(df.ffill(limit=2))

data = pd.Series([1.0, np.nan, 3.5, np.nan, 7])
print(data.fillna(data.mean()))
```

```python
# Removing duplicates
data = pd.DataFrame({"k1": ["one", "two"] * 3 + ["two"], "k2": [1, 1, 2, 3, 3, 4, 4]})
print(data)
print(data.duplicated())
print(data.drop_duplicates())

data["v1"] = range(7)
print(data)
print(data.drop_duplicates(subset=["k1"]))
```

#### Tuple Unpacking (Giải nén tuple)

```python
# Gán nhiều biến cùng lúc
tup = (4, 5, 6)
a, b, c = tup
print(b)  # Output: 5
```

#### Slicing (Cắt list)

```python
seq = [7, 2, 3, 7, 5, 6, 0, 1]
```

```python
# Cắt từ index 1 đến 5 (không bao gồm 5)
print(seq[1:5])  # Output: [2, 3, 7, 5]

# Gán slice với list mới
seq[3:5] = [6, 3]
print(seq)  # Output: [7, 2, 3, 6, 3, 6, 0, 1]

# Slice từ đầu đến index 5
print(seq[:5])   # Output: [7, 2, 3, 6, 3]
# Slice từ index 3 đến cuối
print(seq[3:])   # Output: [6, 3, 6, 0, 1]

# Negative indexing
print(seq[-4:])   # Output: [3, 6, 0, 1] - 4 phần tử cuối
print(seq[-6:-2]) # Output: [6, 3, 6, 0] - từ index -6 đến -2

# Step slicing
print(seq[::2])   # Output: [7, 3, 3, 0] - lấy mỗi phần tử thứ 2
print(seq[::-1])  # Output: [1, 0, 6, 3, 6, 3, 2, 7] - đảo ngược list
```

#### Nhóm dữ liệu theo key

```python
words = ["apple", "bat", "bar", "atom", "book"]

# Cách thủ công
by_letter = {}
for word in words:
    letter = word[0]  # Lấy chữ cái đầu
    if letter not in by_letter:
        by_letter[letter] = [word]
    else:
        by_letter[letter].append(word)
print(by_letter)  # Output: {'a': ['apple', 'atom'], 'b': ['bat', 'bar', 'book']}
```

```python
# Sử dụng setdefault()
by_letter = {}
for word in words:
    letter = word[0]
    # setdefault() trả về value nếu key tồn tại, nếu không thì set value mặc định
    by_letter.setdefault(letter, []).append(word)
print(by_letter)
```

```python
# Sử dụng defaultdict
from collections import defaultdict
by_letter = defaultdict(list)  # Tự động tạo list nếu key chưa tồn tại
for word in words:
    by_letter[word[0]].append(word)
print(by_letter)
```

#### Phép toán Set

```python
a = {1, 2, 3, 4, 5}
b = {3, 4, 5, 6, 7, 8}
```

```python
# Union (hợp): tất cả phần tử của cả hai set
print(a.union(b))    # Output: {1, 2, 3, 4, 5, 6, 7, 8}
print(a | b)         # Toán tử | cũng là union
```

```python
# Intersection (giao): phần tử chung
print(a.intersection(b))  # Output: {3, 4, 5}
print(a & b)             # Toán tử & cũng là intersection
```

```python
# In-place operations
c = a.copy()
c |= b  # Equivalent to c = c.union(b)
print(c)  # Output: {1, 2, 3, 4, 5, 6, 7, 8}

d = a.copy()  
d &= b  # Equivalent to d = d.intersection(b)
print(d)  # Output: {3, 4, 5}
```

#### Nested List Comprehension

```python
all_data = [["John", "Emily", "Michael", "Mary", "Steven"],
            ["Maria", "Juan", "Javier", "Natalia", "Pilar"]]
```

```python
# Cách thủ công
names_of_interest = []
for names in all_data:
    enough_as = [name for name in names if name.count("a") >= 2]
    names_of_interest.extend(enough_as)
print(names_of_interest)  # Output: ['Maria', 'Natalia']
```

```python
# Nested comprehension
result = [name for names in all_data for name in names
          if name.count("a") >= 2]
print(result)  # Output: ['Maria', 'Natalia']
```

#### Flattening với Comprehension

```python
some_tuples = [(1, 2, 3), (4, 5, 6), (7, 8, 9)]

# Flatten tuple thành list
flattened = [x for tup in some_tuples for x in tup]
print(flattened)  # Output: [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
# Equivalent manual approach:
flattened = []
for tup in some_tuples:
    for x in tup:
        flattened.append(x)

# List of lists comprehension
print([[x for x in tup] for tup in some_tuples])
# Output: [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
```

#### Variable Scope

```python
# Sử dụng global keyword để rebind global variable
a = None
def bind_a_variable():
    global a  # Khai báo sử dụng global variable
    a = []

bind_a_variable()
print(a)  # Output: []
```

#### Functions as Objects

```python
states = ["   Alabama ", "Georgia!", "Georgia", "georgia", "FlOrIda",
          "south   carolina##", "West virginia?"]
```

```python
# Function xử lý string
import re

def clean_strings(strings):
    """Function để clean list of strings"""
    result = []
    for value in strings:
        value = value.strip()                    # Loại bỏ whitespace
        value = re.sub("[!#?]", "", value)      # Loại bỏ punctuation
        value = value.title()                   # Capitalize từng từ
        result.append(value)
    return result
```

```python
clean_strings(states)
```

```python
# Refactor: sử dụng list of functions
def remove_punctuation(value):
    return re.sub("[!#?]", "", value)

# List các function để apply
clean_ops = [str.strip, remove_punctuation, str.title]

def clean_strings(strings, ops):
    """Function nhận list of operations"""
    result = []
    for value in strings:
        for func in ops:  # Apply từng function
            value = func(value)
        result.append(value)
    return result
```

```python
clean_strings(states, clean_ops)
```

### Generators (Bộ sinh)

#### Generator Functions

```python
def squares(n=10):
    """Generator function tạo bình phương các số"""
    print(f"Generating squares from 1 to {n ** 2}")
    for i in range(1, n + 1):
        yield i ** 2  # yield thay vì return
```

```python
# Tạo generator object
gen = squares()
print(gen)  # Output: <generator object squares at 0x...>
```

```python
# Iterate qua generator
for x in gen:
    print(x, end=" ")  # Output: 1 4 9 16 25 36 49 64 81 100
```

#### Generator Expressions

```python
# Generator expression
gen = (x ** 2 for x in range(100))
print(gen)  # Output: <generator object <genexpr> at 0x...>
```

```python
# Sử dụng trong function calls
print(sum(x ** 2 for x in range(100)))  # Output: 328350
print(dict((i, i ** 2) for i in range(5)))  # Output: {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
```

#### itertools.groupby

```python
import itertools

def first_letter(x):
    return x[0]

names = ["Alan", "Adam", "Wes", "Will", "Albert", "Steven"]

# groupby(): nhóm các phần tử liên tiếp có cùng key
for letter, names in itertools.groupby(names, first_letter):
    print(letter, list(names))  # names là generator nên cần list()
# Output:
# A ['Alan', 'Adam']
# W ['Wes', 'Will']  
# A ['Albert']
# S ['Steven']
```

### Exception Handling 

#### Basic Exception Handling

```python
# Function có thể raise exception
print(float("1.2345"))  # Output: 1.2345
# float("something")    # ValueError!
```

```python
def attempt_float(x):
    """Safely convert to float"""
    try:
        return float(x)
    except:
        return x  # Trả về original value nếu không convert được
```

```python
print(attempt_float("1.2345"))   # Output: 1.2345
print(attempt_float("something")) # Output: something
```

#### Specific Exception Types

```python
float((1, 2)) # float((1, 2))  # TypeError!
```

```python
# Handle multiple exception types
def attempt_float(x):
    """Handle multiple exception types"""
    try:
        return float(x)
    except (TypeError, ValueError):  # Tuple of exception types
        return x

print(attempt_float((1, 2)))  # Output: (1, 2)
```

- Để đọc file theo từng phần, dùng `chunksize` để chỉ định số lượng hàng mỗi lần đọc.

```python
chunker = pd.read_csv("examples/ex6.csv", chunksize=1000)
type(chunker)
```

- `pandas.read_csv` trả về TextFileReader, cho phép lặp qua các phần của file theo chunksize.

> Ví dụ: có thể lặp qua ex6.csv, tính tổng số lần xuất hiện của các giá trị trong cột "key".

```python
chunker = pd.read_csv("examples/ex6.csv", chunksize=1000)
tot = pd.Series([], dtype='int64')
for piece in chunker:
    tot = tot.add(piece["key"].value_counts(), fill_value=0)
tot = tot.sort_values(ascending=False)
```

### Parsing XML with lxml.objectify

- Sử dụng lxml.objectify để phân tích (parse) file XML.
- Dùng getroot để lấy tham chiếu đến nút gốc (root node) của file XML.

```python
from lxml import objectify

path = "datasets/mta_perf/Performance_MNR.xml"
with open(path) as f:
    parsed = objectify.parse(f)
root = parsed.getroot()
```

- `root.INDICATOR` trả về generator, sinh ra từng phần tử `<INDICATOR>` trong XML.
- Với mỗi bản ghi, có thể tạo dictionary ánh xạ tên thẻ (tag) như YTD_ACTUAL tới giá trị dữ liệu, bỏ qua một số thẻ nhất định.

```python
data = []

skip_fields = ["PARENT_SEQ", "INDICATOR_SEQ",
               "DESIRED_CHANGE", "DECIMAL_PLACES"]

for elt in root.INDICATOR:
    el_data = {}
    for child in elt.getchildren():
        if child.tag in skip_fields:
            continue
        el_data[child.tag] = child.pyval
    data.append(el_data)
```

- Cuối cùng, chuyển danh sách các dictionary thành DataFrame để dễ dàng xử lý và phân tích dữ liệu.

```python
perf = pd.DataFrame(data)
perf.head()
```

- `pandas.read_xml` giúp chuyển XML thành DataFrame nhanh chóng, thay thế cho việc xử lý từng thẻ XML thủ công.

```python
perf2 = pd.read_xml(path)
perf2.head()
```

```python
frame = pd.DataFrame({"a": np.random.standard_normal(100)})
store = pd.HDFStore("examples/mydata.h5")
store["obj1"] = frame
store["obj1_col"] = frame["a"]
store
```

```python
data[4] = np.nan
data
data.dropna(axis="columns", how="all")
```

- Muốn giữ lại các hàng chỉ có tối đa một số lượng NA nhất định, dùng tham số `thresh`.

```python
df = pd.DataFrame(np.random.standard_normal((7, 3)))
df.iloc[:4, 1] = np.nan
df.iloc[:2, 2] = np.nan
df
df.dropna()
df.dropna(thresh=2)
```

### Filling In Missing Data

- Thay vì loại bỏ dữ liệu thiếu, có thể điền giá trị thay thế vào chỗ trống.
- Hàm `fillna` thường được dùng để làm việc này.
- Gọi `fillna` với một hằng số sẽ thay thế mọi giá trị thiếu bằng hằng số đó.

```python
df.fillna(0)
```

- Gọi `fillna` với một **dictionary**, bạn có thể chỉ định giá trị thay thế khác nhau cho từng cột.

```python
df.fillna({1: 0.5, 2: 0})
```

Các phương pháp nội suy (interpolation) có sẵn cho **reindexing** cũng có thể được sử dụng với `fillna`.

```python
df = pd.DataFrame(np.random.standard_normal((6, 3)))
df.iloc[2:, 1] = np.nan
df.iloc[4:, 2] = np.nan
df
df.fillna(method="ffill")
df.fillna(method="ffill", limit=2)
```

- Với `fillna`, có thể thay giá trị thiếu bằng các thống kê đơn giản như **trung vị (median)** hoặc **trung bình (mean)**.

```python
data = pd.Series([1., np.nan, 3.5, np.nan, 7])
data.fillna(data.mean())
```

## 7.2 Data Transformation

- Chương này đến giờ tập trung vào xử lý dữ liệu thiếu.
- Ngoài ra, còn có các thao tác quan trọng khác như lọc, làm sạch và biến đổi dữ liệu.

### Removing Duplicates

- Trong DataFrame có thể xuất hiện các hàng trùng lặp vì nhiều lý do khác nhau.

> Ví dụ như sau:

```python
data = pd.DataFrame({"k1": ["one", "two"] * 3 + ["two"],
                     "k2": [1, 1, 2, 3, 3, 4, 4]})
data
```

- `duplicated` trả về Series Boolean, xác định hàng nào bị trùng với hàng trước đó.

```python
data.duplicated()
```

- `drop_duplicates` trả về DataFrame đã loại bỏ các hàng trùng lặp.

```python
data.drop_duplicates()
```

- Mặc định, cả hai phương thức xem xét tất cả các cột.
- Có thể chọn một số cột (ví dụ: chỉ cột "k1") để phát hiện và lọc trùng lặp.

```python
data["v1"] = range(7)
data
data.drop_duplicates(subset=["k1"])
```

- Mặc định, `duplicated` và `drop_duplicates` giữ giá trị trùng đầu tiên.
- Dùng `keep="last"` để giữ giá trị trùng cuối cùng.

```python
data.drop_duplicates(["k1", "k2"], keep="last")
```

### Transforming Data Using a Function or Mapping

- Thường cần **biến đổi dữ liệu** dựa trên giá trị trong mảng, Series hoặc cột của DataFrame.
- Ví dụ minh họa: dữ liệu giả định về các loại thịt.

```python
data = pd.DataFrame({"food": ["bacon", "pulled pork", "bacon",
                              "pastrami", "corned beef", "bacon",
                              "pastrami", "honey ham", "nova lox"],
                     "ounces": [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

- Có thể thêm một cột mới để chỉ rõ loại động vật mà mỗi loại thịt xuất phát từ đó, bằng cách tạo bảng ánh xạ (mapping) giữa thịt và động vật.

```python
meat_to_animal = {
  "bacon": "pig",
  "pulled pork": "pig",
  "pastrami": "cow",
  "corned beef": "cow",
  "honey ham": "pig",
  "nova lox": "salmon"
}
```

- Phương thức map trên Series cho phép **biến đổi giá trị** bằng cách truyền vào một hàm hoặc một **đối tượng dạng** dictionary để ánh xạ.

```python
data["animal"] = data["food"].map(meat_to_animal)
data
```

- Ta cũng có thể truyền trực tiếp một hàm để thực hiện toàn bộ việc biến đổi giá trị.

```python
def get_animal(x):
    return meat_to_animal[x]
data["food"].map(get_animal)
```

- `map` là cách tiện lợi để thực hiện biến đổi từng phần tử và các thao tác làm sạch dữ liệu.

### Replacing Values

- `fillna` chỉ là trường hợp đặc biệt của việc thay thế giá trị.
- `map` có thể sửa một phần giá trị, nhưng `replace` đơn giản và linh hoạt hơn.

>  Ví dụ minh họa với một Series:

```python
data = pd.Series([1., -999., 2., -999., -1000., 3.])
data
```

- Giá trị -999 có thể là dấu hiệu cho dữ liệu thiếu.
- Dùng `replace` để thay bằng NA mà pandas nhận diện, tạo ra Series mới.

```python
data.replace(-999, np.nan)
```

- Để thay thế từng giá trị khác nhau, truyền một danh sách các giá trị thay thế.

```python
data.replace([-999, -1000], np.nan)
```

- Muốn thay mỗi giá trị bằng một giá trị riêng biệt, truyền vào danh sách các giá trị thay thế.

```python
data.replace([-999, -1000], [np.nan, 0])
```

- Tham số truyền vào cũng có thể là một **dictionary** để chỉ định giá trị thay thế cho từng mục cụ thể.

```python
data.replace({-999: np.nan, -1000: 0})
```

### Renaming Axis Indexes

- Tương tự như giá trị trong Series, **nhãn trục (axis labels)** cũng có thể được biến đổi bằng hàm hoặc mapping.
- Có thể **sửa trực tiếp nhãn trục** mà không tạo đối tượng mới.

```python
data = pd.DataFrame(np.arange(12).reshape((3, 4)),
                    index=["Ohio", "Colorado", "New York"],
                    columns=["one", "two", "three", "four"])
```

- Giống Series, **index của trục (axis)** cũng có phương thức `map`.

```python
def transform(x):
    return x[:4].upper()

data.index.map(transform)
```

- Có thể gán trực tiếp cho thuộc tính index để sửa DataFrame ngay tại chỗ (in place):

```python
data.index = data.index.map(transform)
data
```

- Nếu muốn tạo một phiên bản **biến đổi của dataset mà không thay đổi bản gốc**, có thể dùng phương thức `rename`.

```python
data.rename(index=str.title, columns=str.upper)
```

- Đặc biệt, `rename` có thể dùng với **dictionary**, chỉ định nhãn mới cho một tập con các nhãn trục.

```python
data.rename(index={"OHIO": "INDIANA"},
            columns={"three": "peekaboo"})
```

- `rename` giúp bạn tránh phải sao chép DataFrame và gán nhãn mới thủ công cho **index và columns**.

### Discretization and Binning

- Dữ liệu liên tục thường được chia thành các “bins” để phân tích.

> Ví dụ: nhóm người theo khoảng tuổi rời rạc.

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
```

- Chia dữ liệu thành các khoảng tuổi: 18–25, 26–35, 36–60, 61 trở lên.
- Dùng pandas.cut để thực hiện việc phân nhóm này.

```python
bins = [18, 25, 35, 60, 100]
age_categories = pd.cut(ages, bins)
age_categories
```

- Kết quả từ `pandas.cut` là một **đối tượng Categorical** đặc biệt.
- Mỗi bin được đại diện bởi một interval chứa giới hạn dưới và trên của bin.

```python
age_categories.codes
age_categories.categories
age_categories.categories[0]
pd.value_counts(age_categories)
```

- `pd.value_counts(categories)` cho số lượng phần tử trong mỗi bin của `pandas.cut`.
- Trong biểu diễn chuỗi của interval:
    - `( hoặc )` → mở (exclusive)
    - `[ hoặc ]` → đóng (inclusive)

- Có thể thay đổi phía đóng bằng `right=False`.

```python
pd.cut(ages, bins, right=False)
```

- Có thể thay nhãn mặc định của các bin bằng cách truyền một danh sách hoặc mảng vào tùy chọn labels.

```python
group_names = ["Youth", "YoungAdult", "MiddleAged", "Senior"]
pd.cut(ages, bins, labels=group_names)
```

- Nếu truyền một số nguyên cho `pandas.cut` thay vì các giới hạn bin cụ thể, pandas sẽ tạo các bin có độ dài bằng nhau dựa trên giá trị min và max của dữ liệu.

> Ví dụ: dữ liệu phân phối đều được chia thành 4 phần.

```python
data = np.random.uniform(size=20)
pd.cut(data, 4, precision=2)
```

- `precision=2` giới hạn độ chính xác thập phân trong nhãn bin.
- `pandas.qcut` chia dữ liệu dựa trên quantile, tạo các bin có số phần tử xấp xỉ bằng nhau, khác với `pandas.cut` dùng khoảng giá trị.

```python
data = np.random.standard_normal(1000)
quartiles = pd.qcut(data, 4, precision=2)
quartiles
pd.value_counts(quartiles)
```

- Tương tự pandas.cut, bạn có thể truyền quantile tùy chỉnh (giá trị từ 0 đến 1) cho pandas.qcut.

```python
pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.]).value_counts()
```

### Detecting and Filtering Outliers

### Computing Indicator/Dummy Variables

- Một dạng biến đổi khác cho mô hình thống kê hoặc học máy là **chuyển biến categorical thành ma trận dummy/indicator**.
- Nếu một cột có **k giá trị khác nhau**, tạo DataFrame với **k cột chứa 1 và 0**.
- Pandas có hàm `pandas.get_dummies` để làm việc này, hoặc có thể tự viết hàm.

> Ví dụ minh họa với một DataFrame.

```python
df = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "b"],
                   "data1": range(6)})
df
pd.get_dummies(df["key"], dtype=float)
```

- Trong một số trường hợp, bạn muốn thêm tiền tố (prefix) cho các cột trong DataFrame indicator trước khi gộp với dữ liệu khác.
- Hàm `pandas.get_dummies` có tham số prefix để thực hiện việc này.

```python
dummies = pd.get_dummies(df["key"], prefix="key", dtype=float)
df_with_dummy = df[["data1"]].join(dummies)
df_with_dummy
```

- Phương thức DataFrame.join sẽ được giải thích chi tiết ở chương sau.
- Nếu một hàng thuộc nhiều category, cần dùng cách khác để tạo biến dummy.

> Ví dụ: bộ dữ liệu MovieLens 1M, được phân tích chi tiết ở Chương 13.

```python
mnames = ["movie_id", "title", "genres"]
movies = pd.read_table("datasets/movielens/movies.dat", sep="::",
                       header=None, names=mnames, engine="python")
movies[:10]
```

- Pandas có phương thức đặc biệt Series.str.get_dummies (các phương thức bắt đầu bằng str. sẽ được giải thích chi tiết ở mục 7.4 “String Manipulation”).
- Phương thức này xử lý trường hợp một hàng thuộc nhiều nhóm, được mã hóa dưới dạng chuỗi phân tách.

```python
dummies = movies["genres"].str.get_dummies("|")
dummies.iloc[:10, :6]
```

- Sau đó, giống như trước, bạn có thể gộp với DataFrame movies và thêm tiền tố `"Genre_"` vào tên cột trong DataFrame dummy bằng `add_prefix`.

```python
movies_windic = movies.join(dummies.add_prefix("Genre_"))
movies_windic.iloc[0]
```

- Một cách hữu ích trong thống kê là kết hợp `pandas.get_dummies` với hàm **discretization** như `pandas.cut`.

```python
np.random.seed(12345) # to make the example repeatable
values = np.random.uniform(size=10)
values
bins = [0, 0.2, 0.4, 0.6, 0.8, 1]
pd.get_dummies(pd.cut(values, bins))
```

## 7.4 String Manipulation

- Python từ lâu đã phổ biến cho xử lý dữ liệu thô, đặc biệt là chuỗi và văn bản, nhờ các phương thức sẵn có của string.
- Với các thao tác phức tạp hơn cần regular expressions.
- Pandas giúp áp dụng chuỗi và regex trên toàn bộ mảng dữ liệu một cách gọn gàng, đồng thời xử lý dữ liệu thiếu.

### Python Built-In String Object Methods

- Trong nhiều ứng dụng xử lý chuỗi, các phương thức string sẵn có là đủ.

> Ví dụ: chuỗi phân tách bằng dấu phẩy có thể tách thành các phần bằng `split`.

```python
val = "a,b,  guido"
val.split(",")
```

- `split` thường được kết hợp với `strip` để loại bỏ dấu cách thừa (bao gồm xuống dòng).

```python
pieces = [x.strip() for x in val.split(",")]
pieces
```

- Các **substring** này có thể được nối lại với delimiter `::` bằng phép cộng `(+)`.

```python
first, second, third = pieces
first + "::" + second + "::" + third
```

- Tuy nhiên, cách này không phù hợp tổng quát.
- Cách nhanh hơn và Pythonic là truyền danh sách hoặc tuple vào phương thức `"::".join()`.

```python
"::".join(pieces)
```

- Các phương thức khác liên quan đến tìm vị trí substring.
- Dùng in là cách tốt nhất để kiểm tra substring, ngoài ra còn có `index` và `find`.

```python
"guido" in val
val.index(",")
val.find(":")
```

> Lưu ý: `index` sẽ báo lỗi nếu không tìm thấy chuỗi, còn `find` trả về -1.

```python
val.index(":")
```

- Ngoài ra, count trả về số lần xuất hiện của một substring cụ thể.

```python
val.count(",")
```

- `replace` thay thế các mẫu (pattern) bằng mẫu khác.
- Cũng thường dùng để xóa pattern bằng cách truyền chuỗi rỗng.

```python
val.replace(",", "::")
val.replace(",", "")
```

### Regular Expressions

- Regular expressions (regex) cung cấp cách linh hoạt để tìm kiếm hoặc khớp các mẫu chuỗi, thường là phức tạp.
- Python dùng module re để áp dụng regex cho chuỗi.
- Các hàm trong re thuộc ba nhóm: pattern matching, substitution, và splitting.

> Ví dụ: muốn tách chuỗi theo số lượng khoảng trắng biến thiên (tabs, spaces, newlines), dùng regex \s+.

```python
import re
text = "foo    bar\t baz  \tqux"
re.split(r"\s+", text)
```

- Khi gọi `re.split(r"\s+", text)`, regex sẽ được biên dịch trước, sau đó dùng phương thức split trên chuỗi.
- Có thể tự biên dịch regex bằng `re.compile` để tạo regex object có thể tái sử dụng.

```python
regex = re.compile(r"\s+")
regex.split(text)
```

- Nếu muốn lấy danh sách tất cả các mẫu khớp với regex, dùng phương thức `findall`.

```python
regex.findall(text)
```

- Nên dùng `re.compile` khi áp dụng cùng một regex cho nhiều chuỗi để tiết kiệm CPU.
- `match` và `search` liên quan đến `findall`:
    - `findall` trả về tất cả khớp.
    - `search` trả về khớp đầu tiên.
    - `match` chỉ khớp tại đầu chuỗi.

> Ví dụ: regex để nhận diện email trong một đoạn văn bản.

```python
text = """Thuong thuong@google.com
Thi thi@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com"""
pattern = r"[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}"

# re.IGNORECASE makes the regex case insensitive
regex = re.compile(pattern, flags=re.IGNORECASE)
```

- Dùng `findall` trên văn bản sẽ trả về danh sách các địa chỉ email:

```python
regex.findall(text)
```

- `search` trả về một match object đặc biệt cho email đầu tiên trong văn bản.
- Match object chỉ cho biết vị trí bắt đầu và kết thúc của pattern trong chuỗi.

```python
m = regex.search(text)
m
text[m.start():m.end()]
```

- `regex.match` trả về None, vì nó chỉ khớp nếu pattern xuất hiện tại đầu chuỗi.

```python
print(regex.match(text))
```

- Tương tự, sub trả về một chuỗi mới với các pattern được thay bằng chuỗi khác.

```python
print(regex.sub("REDACTED", text))
```

- Giả sử muốn tìm email và chia mỗi email thành ba phần: username, domain name, domain suffix.
- Cách làm: đặt ngoặc quanh các phần trong pattern để tách nhóm.

```python
pattern = r"([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})"
regex = re.compile(pattern, flags=re.IGNORECASE)
```

- Match object từ regex đã chỉnh sửa sẽ trả về một tuple các thành phần của pattern bằng phương thức `groups`.

```python
m = regex.match("wesm@bright.net")
m.groups()
```

- `findall` trả về một danh sách các tuple nếu pattern có các nhóm (groups).

```python
regex.findall(text)
```

- `sub` cũng có thể truy cập các group trong mỗi match bằng các ký hiệu đặc biệt như `\1`, `\2`:
    - `\1` → nhóm đầu tiên
    - `\2` → nhóm thứ hai
    - v.v.

```python
print(regex.sub(r"Username: \1, Domain: \2, Suffix: \3", text))
```

### Categorical Extension Type in pandas

- Pandas có extension type Categorical dùng integer-based encoding.
- Đây là kỹ thuật nén dữ liệu phổ biến cho dữ liệu có nhiều giá trị lặp lại, giúp tăng tốc độ và tiết kiệm bộ nhớ, đặc biệt với dữ liệu chuỗi.

> Ví dụ: Series đã xem trước đó.

```python
fruits = ['apple', 'orange', 'apple', 'apple'] * 2
N = len(fruits)
rng = np.random.default_rng(seed=12345)
df = pd.DataFrame({'fruit': fruits,
                   'basket_id': np.arange(N),
                   'count': rng.integers(3, 15, size=N),
                   'weight': rng.uniform(0, 4, size=N)},
                  columns=['basket_id', 'fruit', 'count', 'weight'])
df
```

- Ở đây, `df['fruit']` là mảng các string Python.

- Giá trị của `fruit_cat` giờ là `pandas.Categorical`, có thể truy cập qua thuộc tính `.array`.

```python
c = fruit_cat.array
type(c)
```

- Categorical object có hai thuộc tính: `categories` và `codes`.

```python
c.categories
c.codes
```

> Một mẹo hữu ích để lấy mapping giữa codes và categories là:

```python
dict(enumerate(c.categories))
```

- Có thể chuyển cột DataFrame sang categorical bằng cách gán kết quả đã chuyển đổi.

```python
df['fruit'] = df['fruit'].astype('category')
df["fruit"]
```

- Cũng có thể tạo **pandas.Categorical** trực tiếp từ các sequence khác của Python.

```python
my_categories = pd.Categorical(['foo', 'bar', 'baz', 'foo', 'bar'])
my_categories
```

- Nếu đã có dữ liệu categorical được mã hóa từ nguồn khác, có thể dùng `from_codes` constructor.

```python
categories = ['foo', 'bar', 'baz']
codes = [0, 1, 2, 0, 0, 1]
my_cats_2 = pd.Categorical.from_codes(codes, categories)
my_cats_2
```

- Nếu không chỉ định, các categorical conversion không giả định thứ tự cụ thể của categories.
- Mảng categories có thể khác nhau tùy dữ liệu đầu vào.
- Khi dùng `from_codes` hoặc các constructor khác, có thể chỉ định thứ tự có ý nghĩa cho categories.

```python
ordered_cat = pd.Categorical.from_codes(codes, categories,
                                        ordered=True)
ordered_cat
```

- Kết quả `[foo < bar < baz]` nghĩa là `'foo'` đứng trước `'bar'`,…
- Một categorical không có thứ tự có thể được chuyển thành ordered bằng as_ordered.

```python
my_cats_2.as_ordered()
```

### Computations with Categoricals

- Dùng Categorical trong pandas so với phiên bản không mã hóa (mảng string) về cơ bản giống nhau.
- Một số phần của pandas, như groupby, chạy nhanh hơn với categorical.
- Một số hàm khác có thể dùng ordered flag.

> Ví dụ: dùng `pandas.qcut` để bin dữ liệu số ngẫu nhiên; hàm này trả về `pandas.Categorical`.

```python
rng = np.random.default_rng(seed=12345)
draws = rng.standard_normal(1000)
draws[:5]
```

- Thực hiện chia dữ liệu thành các quartile và trích xuất thống kê.

```python
bins = pd.qcut(draws, 4)
bins
```

- Mặc dù hữu ích, quartile thực tế có thể ít trực quan khi báo cáo.
- Có thể dùng labels trong qcut để đặt tên cho các quartile.

```python
bins = pd.qcut(draws, 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
bins
bins.codes[:10]
```

- Categorical với nhãn không chứa thông tin về biên của bin trong dữ liệu.
- Có thể dùng `groupby` để trích xuất thống kê tóm tắt.

```python
bins = pd.Series(bins, name='quartile')
results = (pd.Series(draws)
           .groupby(bins)
           .agg(['count', 'min', 'max'])
           .reset_index())
results
```

- Cột 'quartile' trong kết quả giữ thông tin categorical gốc, bao gồm cả thứ tự, từ bins.

```python
results['quartile']
```

## Nối DataFrame theo kiểu cơ sở dữ liệu
### Các thao tác hợp nhất hoặc nối kết hợp các tập dữ liệu bằng cách liên kết các hàng bằng một hoặc nhiều khóa. Các thao tác này đặc biệt quan trọng trong cơ sở dữ liệu quan hệ (ví dụ: dựa trên SQL).
### Hàm pandas.merge trong pandas là điểm khởi đầu chính để sử dụng các thuật toán này trên dữ liệu của bạn. 
### Hãy bắt đầu với một ví dụ đơn giản:

```python
df1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "a", "b"],
                    "data1": pd.Series(range(7), dtype="Int64")})
df2 = pd.DataFrame({"key": ["a", "b", "d"],
                    "data2": pd.Series(range(3), dtype="Int64")})
df1
df2
```

### Đây là một ví dụ về phép nối nhiều-một; dữ liệu trong df1 có nhiều hàng được gắn nhãn a và b, trong khi df2 chỉ có một hàng cho mỗi giá trị trong cột khóa. Gọi pandas.merge với các đối tượng này, chúng ta thu được:

```python
pd.merge(df1, df2)
```

### Lưu ý rằng tôi không chỉ định cột nào sẽ được nối. Nếu thông tin đó không được chỉ định, pandas.merge sẽ sử dụng các tên cột chồng chéo làm khóa. Tuy nhiên, tốt hơn hết là nên chỉ định rõ ràng:

```python
pd.merge(df1, df2, on="key")
```

```python
df3 = pd.DataFrame({"lkey": ["b", "b", "a", "c", "a", "a", "b"],
                    "data1": pd.Series(range(7), dtype="Int64")})
df4 = pd.DataFrame({"rkey": ["a", "b", "d"],
                    "data2": pd.Series(range(3), dtype="Int64")})
pd.merge(df3, df4, left_on="lkey", right_on="rkey")
```

### Bạn có thể nhận thấy các giá trị "c" và "d" cùng dữ liệu liên quan bị thiếu trong kết quả. Theo mặc định, pandas.merge thực hiện phép nối "bên trong"; các khóa trong kết quả là giao điểm, hoặc tập hợp chung được tìm thấy trong cả hai bảng. Các tùy chọn khả thi khác là "trái", "phải" và "ngoài". Phép nối ngoài lấy hợp của các khóa, kết hợp hiệu ứng của việc áp dụng cả phép nối trái và phải:

```python
pd.merge(df1, df2, how="outer")
pd.merge(df3, df4, left_on="lkey", right_on="rkey", how="outer")
```

```python
df1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "b"],
                    "data1": pd.Series(range(6), dtype="Int64")})
df2 = pd.DataFrame({"key": ["a", "b", "a", "b", "d"],
                    "data2": pd.Series(range(5), dtype="Int64")})
df1
df2
pd.merge(df1, df2, on="key", how="left")
```

```python
pd.merge(df1, df2, how="inner")
```

```python
left = pd.DataFrame({"key1": ["foo", "foo", "bar"],
                     "key2": ["one", "two", "one"],
                     "lval": pd.Series([1, 2, 3], dtype='Int64')})
right = pd.DataFrame({"key1": ["foo", "foo", "bar", "bar"],
                      "key2": ["one", "one", "one", "two"],
                      "rval": pd.Series([4, 5, 6, 7], dtype='Int64')})
pd.merge(left, right, on=["key1", "key2"], how="outer")
```

### Khi bạn nối các cột với nhau, các chỉ mục trên các đối tượng DataFrame đã truyền sẽ bị loại bỏ. Nếu bạn cần giữ nguyên các giá trị chỉ mục, bạn có thể sử dụng reset_index để thêm chỉ mục vào các cột.

```python
pd.merge(left, right, on="key1")
```

```python
pd.merge(left, right, on="key1", suffixes=("_left", "_right"))
```

### Hợp nhất trên Chỉ mục
### Trong một số trường hợp, khóa hợp nhất trong DataFrame sẽ được tìm thấy trong chỉ mục (nhãn hàng) của nó. Trong trường hợp này, bạn có thể truyền left_index=True hoặc right_index=True (hoặc cả hai) để chỉ định rằng chỉ mục nên được sử dụng làm khóa hợp nhất:

```python
left1 = pd.DataFrame({"key": ["a", "b", "a", "a", "b", "c"],
                      "value": pd.Series(range(6), dtype="Int64")})
right1 = pd.DataFrame({"group_val": [3.5, 7]}, index=["a", "b"])
left1
right1
pd.merge(left1, right1, left_on="key", right_index=True)
```

### Nếu bạn xem xét kỹ ở đây, bạn sẽ thấy các giá trị chỉ mục cho left1 đã được giữ nguyên, trong khi ở các ví dụ khác ở trên, các chỉ mục của các đối tượng DataFrame đầu vào bị loại bỏ. Vì chỉ mục của right1 là duy nhất, phép hợp nhất "nhiều-một" này (với phương thức how="inner" mặc định) có thể giữ nguyên các giá trị chỉ mục từ left1 tương ứng với các hàng trong đầu ra.

```python
pd.merge(left1, right1, left_on="key", right_index=True, how="outer")
```

### 24. Random sampling và subsets

```python
suits = ["H", "S", "C", "D"]  # Hearts, Spades, Clubs, Diamonds # Các chất bài
card_val = (list(range(1, 11)) + [10] * 3) * 4 # Giá trị của các lá bài
base_names = ["A"] + list(range(2, 11)) + ["J", "K", "Q"] # Tên các lá bài
cards = [] # Danh sách các lá bài
for suit in suits:
    cards.extend(str(num) + suit for num in base_names) # Tạo tên lá bài với chất tương ứng

deck = pd.Series(card_val, index=cards) # Tạo Series đại diện cho bộ bài
```

### 26. Correlation và regression

```python
close_px = pd.read_csv("examples/stock_px.csv", parse_dates=True,
                       index_col=0) # Đọc dữ liệu từ file CSV với cột đầu tiên làm index và phân tích ngày tháng
close_px.info() # Hiển thị thông tin về DataFrame
close_px.tail(4) # Hiển thị 4 dòng cuối của DataFrame
```

```python
def spx_corr(group):
    return group.corrwith(group["SPX"]) # Hàm tính hệ số tương quan với cột SPX
```

```python
rets = close_px.pct_change().dropna() # Tính lợi suất hàng ngày và loại bỏ NaN
```

```python
def get_year(x):
    return x.year # Hàm trả về năm từ đối tượng datetime

by_year = rets.groupby(get_year) # Nhóm theo năm
by_year.apply(spx_corr) # Áp dụng hàm spx_corr cho mỗi nhóm
```

```python
def corr_aapl_msft(group):
    return group["AAPL"].corr(group["MSFT"]) # Hàm tính hệ số tương quan giữa AAPL và MSFT
by_year.apply(corr_aapl_msft) # Áp dụng hàm corr_aapl_msft cho mỗi nhóm
```

```python
import statsmodels.api as sm
def regress(data, yvar=None, xvars=None): # Hàm thực hiện hồi quy tuyến tính
    Y = data[yvar] # Biến phụ thuộc
    X = data[xvars] # Ma trận các biến độc lập
    X["intercept"] = 1. # Thêm cột intercept vào ma trận X
    result = sm.OLS(Y, X).fit() # Thực hiện hồi quy tuyến tính
    return result.params # Trả về các hệ số hồi quy
```

```python
by_year.apply(regress, yvar="AAPL", xvars=["SPX"]) # Áp dụng hàm regress cho mỗi nhóm với AAPL là biến phụ thuộc và SPX là biến độc lập
```

## 12.3 Giới thiệu về statsmodels

### Ước tính các mô hình tuyến tính

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# To make the example reproducible
rng = np.random.default_rng(seed=12345)

def dnorm(mean, variance, size=1):
    if isinstance(size, int):
        size = size,
    return mean + np.sqrt(variance) * rng.standard_normal(*size)

N = 100
X = np.c_[dnorm(0, 0.4, size=N),
          dnorm(0, 0.6, size=N),
          dnorm(0, 0.2, size=N)]
eps = dnorm(0, 0.1, size=N)
beta = [0.1, 0.3, 0.5]

y = np.dot(X, beta) + eps
```

```python
X[:5]
y[:5]
```

```python
# Hàm sm.add_constant có thể thêm một cột chặn vào một ma trận hiện có
X_model = sm.add_constant(X)
X_model[:5]
```

```python
# Lớp sm.OLS có thể phù hợp với hồi quy tuyến tính bình phương nhỏ thông thường:
model = sm.OLS(y, X)
```

```python
results = model.fit()
results.params
```

```python
print(results.summary())
```

```python
# Giả sử tất cả các tham số mô hình đều nằm trong một DataFrame:
data = pd.DataFrame(X, columns=['col0', 'col1', 'col2'])
data['y'] = y
data[:5]
```

```python
# Sử dụng API công thức statsmodels và chuỗi công thức Patsy:
results = smf.ols('y ~ col0 + col1 + col2', data=data).fit()
results.params
results.tvalues
```

```python
results.predict(data[:5])
```

### Ước tính các quy trình chuỗi thời gian

```python
init_x = 4

values = [init_x, init_x]
N = 1000

b0 = 0.8
b1 = -0.4
noise = dnorm(0, 0.1, N)
for i in range(N):
    new_x = values[-1] * b0 + values[-2] * b1 + noise[i]
    values.append(new_x)
```

Dữ liệu này có cấu trúc AR(2) (hai độ trễ) với các tham số 0,8 và –0,4. Khi bạn điều chỉnh mô hình AR, bạn có thể không biết số lượng các thuật ngữ bị trễ cần đưa vào, vì vậy bạn có thể điều chỉnh mô hình với số độ trễ lớn hơn:

```python
from statsmodels.tsa.ar_model import AutoReg
MAXLAGS = 5
model = AutoReg(values, MAXLAGS)
results = model.fit()
```
